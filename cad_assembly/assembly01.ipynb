```python
from utils.curation import curate_asset
```

```python
import os
import shutil

ASSETS = [
    "hab_mod",
    "lander",
    "moon",
    "pad",
    "rover/2"
]

BASE_DIR = "assets"


for asset in ASSETS:

    asset_dir = os.path.join(BASE_DIR, asset)

    # find the .usda file
    usda_file = None
    for f in os.listdir(asset_dir):
        if f.endswith(".usda"):
            usda_file = f
            break

    if usda_file is None:
        raise RuntimeError(f"No USDA found in {asset_dir}")

    INPUT_USDA = os.path.join(asset_dir, usda_file)

    # curated directory
    OUTPUT_DIR = os.path.join(asset_dir, "curated")

    # asset name (used for output filenames)
    ASSET_NAME = os.path.splitext(usda_file)[0]

    print("Curating:", ASSET_NAME)

    curate_asset(
        input_usda=INPUT_USDA,
        out_dir=OUTPUT_DIR,
        name=ASSET_NAME
    )

    # copy texture folder "0"
    src_tex = os.path.join(asset_dir, "0")
    dst_tex = os.path.join(OUTPUT_DIR, "0")

    if os.path.exists(src_tex):

        if os.path.exists(dst_tex):
            shutil.rmtree(dst_tex)

        shutil.copytree(src_tex, dst_tex)

        print("Copied textures:", src_tex, "->", dst_tex)

    print("Done\n")
```
