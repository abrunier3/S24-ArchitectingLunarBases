```python
import sys, os
sys.path.insert(0, "/home/ang/usd_ang/lib/python")
os.environ["LD_LIBRARY_PATH"] = "/home/ang/usd_ang/lib"
from pxr import Usd, UsdGeom, UsdShade

FILE = "rover2.usda"

stage = Usd.Stage.Open(FILE)

# ---------------------------------------------------------
# Find the first material in the file
# ---------------------------------------------------------
material = None

for prim in stage.Traverse():
    if prim.IsA(UsdShade.Material):
        material = UsdShade.Material(prim)
        break

if material is None:
    raise RuntimeError("No material found in rover2.usda")

print("Material found:", material.GetPath())

# ---------------------------------------------------------
# Bind material to all meshes
# ---------------------------------------------------------
mesh_count = 0

for prim in stage.Traverse():

    if prim.IsA(UsdGeom.Mesh):

        mesh_count += 1
        print("Binding material to:", prim.GetPath())

        binding = UsdShade.MaterialBindingAPI.Apply(prim)
        binding.Bind(material)

print("Meshes processed:", mesh_count)

stage.GetRootLayer().Save()

print("Material binding complete.")
```
