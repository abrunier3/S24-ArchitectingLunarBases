```python
import sys, os
sys.path.insert(0, "/home/ang/usd_ang/lib/python")
os.environ["LD_LIBRARY_PATH"] = "/home/ang/usd_ang/lib"
from pxr import Usd, UsdGeom, UsdShade

FILE = "pad.usda"

stage = Usd.Stage.Open(FILE)

# ---------------------------------------------------
# Find all materials in the file
# ---------------------------------------------------
materials = []

for prim in stage.Traverse():
    if prim.IsA(UsdShade.Material):
        materials.append(UsdShade.Material(prim))

if not materials:
    raise RuntimeError("No materials found in pad.usda")

print("Materials found:")
for m in materials:
    print(" ", m.GetPath())

material = materials[0]  # use first material

print("\nUsing material:", material.GetPath())

# ---------------------------------------------------
# Bind material to every mesh
# ---------------------------------------------------
mesh_count = 0

for prim in stage.Traverse():

    if prim.IsA(UsdGeom.Mesh):

        mesh_count += 1
        print("Binding material to:", prim.GetPath())

        binding = UsdShade.MaterialBindingAPI.Apply(prim)
        binding.Bind(material)

print("\nMeshes processed:", mesh_count)

stage.GetRootLayer().Save()

print("Material binding completed.")
```
